```python
# This script generates synthetic Activity Points (in the format produced by AirSage). 
# Supply the US county bounding boxes, totla number of point sin each county and sampling fraction.

import random
from shapely.geometry import Polygon, Point
import re
import csv
import pandas as pd
import hashlib
from random import randint
import time
import datetime

#"statefips","countyfips","countyns10","geoid10","name","namelsad10","boundingbox","corner_sw","corner_nw","corner_ne","corner_se","corner_wtf","extentn","extents","extente","extentw"
county = csv.reader(open(r'./bounding_file.csv','r'))

#FIPS,State,County,Avg_Daily_Counts
df_totals = pd.read_csv('./AP_control_totals.csv', index_col='FIPS', header=1, names=['FIPS', 'State', 'County', 'Avg_Daily_Counts'])
#head(df_totals)
#df_totals.ix[2230]

sample_frac = 0.1 #Fraction of random data needed

ap_data = csv.writer(open('./generate_random_ap.csv',"wb"))
ap_data.writerow(["APID", "EMIN", "Date", "County_ID", "Epoch", "Duration", "PointType", "Lat", "Lon", "BlockID", 
                  "DeviceBirthDay", "HomeBlock","WorkBlock","DaysHome","DaysNight", "DaysTotal"])

def get_random_point_in_polygon(poly):
    (minx, miny, maxx, maxy) = poly.bounds
    while True:
        p = Point(random.uniform(minx, maxx), random.uniform(miny, maxy))
        if poly.contains(p):
            return p

counter = 0
for line in county:
    #print line
    if "statefips" in line:
        continue
    p= re.split('[,\(\)]',line[6])
    a=float(re.sub(r'\s', ",", p[2]).split(",")[0]),float(re.sub(r'\s', ",", p[2]).split(",")[1])
    b=float(re.sub(r'\s', ",", p[3]).split(",")[0]),float(re.sub(r'\s', ",", p[3]).split(",")[1])
    c=float(re.sub(r'\s', ",", p[4]).split(",")[0]),float(re.sub(r'\s', ",", p[4]).split(",")[1])
    d=float(re.sub(r'\s', ",", p[5]).split(",")[0]),float(re.sub(r'\s', ",", p[5]).split(",")[1])
    e=float(re.sub(r'\s', ",", p[6]).split(",")[0]),float(re.sub(r'\s', ",", p[6]).split(",")[1])
    pol = ([a,b,c,d,e])
    county_id = int(line[3]) #geoid
    print county_id
    try:
        n = int(round(df_totals.ix[county_id].Avg_Daily_Counts*sample_frac,0)) # number of points to be selected
    except KeyError:
        continue
    
    for i in range(counter,counter + n):
        aps = get_random_point_in_polygon(Polygon(pol))
        Lon = round(aps.x,5)
        Lat = round(aps.y,5)
        EMIN = hashlib.sha1(str(county_id+randint(0,n/5))).hexdigest()
        Epoch = int(time.mktime(datetime.datetime.now().timetuple()))
        Date = time.strftime('%Y-%m-%d', time.localtime(Epoch))
        Duration = random.sample([0, 5, 10, 15, 30],  1)[0] 
        PointType = random.sample(["TP", "EP", "UP", "HP", "WP","HWP"],  1)[0] 
        BlockID = str(county_id) + '00000000000'
        DeviceBirthDay = random.randint(1, 30)
        HomeBlock = str(county_id) + '00000000000'
        WorkBlock = str(county_id) + '00000000000'
        DaysHome = random.randint(1, 30)
        DaysNight =random.randint(1, 30)
        DaysTotal =random.randint(1, 30)
                 
        output = [i, EMIN, Date, county_id, Epoch, Duration, PointType, Lat, Lon, BlockID, DeviceBirthDay, HomeBlock,WorkBlock,DaysHome,DaysNight, DaysTotal]
        ap_data.writerow(output)
        
    counter = counter + n
```
